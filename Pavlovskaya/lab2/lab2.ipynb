```python
from tabulate import tabulate
import random

def randomProbability(probability):
    randomNumber = random.random()
    if randomNumber < probability:
        return 0
    else:
        return 1

def Alex(myMoves, opMoves):
    return 1

def Bob(myMoves, opMoves):
    return 0

def Clara(myMoves, opMoves):
    l = len(opMoves)
    if l == 0:
        return 0
    return opMoves[l-1]

def Denis(myMoves, opMoves):
    l = len(opMoves)
    if l == 0:
        return 0
    move = (opMoves[l-1] + 1) % 2
    return move

def Emma(myMoves, opMoves):
    game = len(opMoves) + 1
    if game % 20 == 0:
        return 1
    return 0

def Frida(myMoves, opMoves):
    if 1 in opMoves:
        return 1
    return 0

def George(myMoves, opMoves):
    l = len(opMoves)
    if l == 0 or l == 1:
        return 0
    last = opMoves[l-1]
    preLast = opMoves[l-2]
    if last == 1 and preLast == 1:
        return 1
    return 0

def Hank(myMoves, opMoves):
    return randomProbability(0.5)

def Ivan(myMoves, opMoves):
    return randomProbability(0.9)
 
def Jack(myMoves, opMoves):
    l = len(myMoves)
    if l == 0:
        return 0
    last = opMoves[l-1]
    if last == 0:
        return 0
    return randomProbability(0.25)
     
def Kevin(myMoves, opMoves):
    l = len(myMoves)
    if l == 0:
        return 0
    last = opMoves[l-1]
    prob = randomProbability(0.25)
    if prob == 1:
        return last
    return (last + 1) % 2 

def Lucas(myMoves, opMoves):
    l = len(myMoves)
    if l == 0:
        Lucas.numbMoves = random.randint(1, 50)
        Lucas.currentMove = 0
        return 0
    Lucas.currentMove += 1
    if Lucas.currentMove % Lucas.numbMoves == 0:
        return 1
    else:
        return 0

def Max(myMoves, opMoves):
    l = len(myMoves)
    if l == 0:
        Max.zeroMoves = random.randint(0, 20)
        Max.oneMoves = random.randint(0, 20)
        Max.moveCount = 0
        Max.currentState = 0
        return 0
    Max.moveCount += 1
    if Max.currentState == 0:
        if Max.moveCount <= Max.zeroMoves:
            return 0
        else:
            Max.currentState = 1
            Max.moveCount = 1
            Max.oneMoves = random.randint(0, 20)
            return 1
    else:
        if Max.moveCount <= Max.oneMoves:
            return 1
        else:
            Max.currentState = 0
            Max.moveCount = 1
            Max.zeroMoves = random.randint(0, 20)
            return 0

def Natan(myMoves, opMoves):
    l = len(opMoves)
    if l == 0:
        return randomProbability(0.5)
    last = opMoves[l-1]
    if randomProbability(0.7) == 1:
        return last
    else:
        return (last + 1) % 2

def reward(f, s):
    if f == 0 and s == 0:
        return 3, 3
    if f == 0 and s == 1:
        return 0, 5
    if f == 1 and s == 0:
        return 5, 0
    if f == 1 and s == 1:
        return 1, 1

points = {
    Alex: 0,
    Bob: 0, 
    Clara: 0,
    Denis: 0,
    Emma: 0,
    Frida: 0,
    George: 0,
    Hank: 0,
    Ivan: 0,
    Jack: 0,
    Kevin: 0,
    Lucas: 0,
    Max: 0,
    Natan: 0
}

series = {
    Alex: 0,
    Bob: 0, 
    Clara: 0,
    Denis: 0,
    Emma: 0,
    Frida: 0,
    George: 0,
    Hank: 0,
    Ivan: 0,
    Jack: 0,
    Kevin: 0,
    Lucas: 0,
    Max: 0,
    Natan: 0,
}

strategies = [Alex, Bob, Clara, Denis, Emma, Frida, George, Hank, Ivan, Jack, Kevin, Lucas, Max, Natan]

stochasticStrategies = [Hank, Ivan, Jack, Kevin, Lucas, Max, Natan]

strategyNames = [s.__name__ for s in strategies]

points = {s: 0 for s in strategies}
serie = {s: 0 for s in strategies}

matrixPoints = {s1: {s2: 0 for s2 in strategies} for s1 in strategies}
matrixSeries = {s1: {s2: 0 for s2 in strategies} for s1 in strategies}

def makeGame(chel1, chel2):
    global points
    global serie

    isStochastic = chel1 in stochasticStrategies or chel2 in stochasticStrategies
    numGames = 1000 if isStochastic else 1

    results1 = []
    results2 = []
    series1_list = []
    series2_list = []

    totalRes1 = 0
    totalRes2 = 0
    maxSer1 = 0
    maxSer2 = 0
    for j in range(numGames):
        cSer1 = 0
        cSer2 = 0
        max_cSer1 = 0 
        max_cSer2 = 0
        points1 = []
        points2 = []
        res1 = 0
        res2 = 0
        
        for i in range(200):
            fst = chel1(points1, points2)
            points1.append(fst)
            sec = chel2(points2, points1)
            points2.append(sec)
            rew1, rew2 = reward(fst, sec)
            res1 += rew1
            res2 += rew2
    
            if rew1 == 5:
                cSer1 += 1
                if cSer1 > max_cSer1:
                    serie[chel1] = cSer1
            else:
                cSer1 = 0
      
            if rew2 == 5:
                cSer2 += 1
                if cSer2 > max_cSer2:
                    serie[chel2] = cSer2
            else:
                cSer2 = 0
        totalRes1 += res1
        totalRes2 += res2
        results1.append(res1) 
        results2.append(res2)
        series1_list.append(max_cSer1)
        series2_list.append(max_cSer2)

    if not isStochastic:
        totalRes1 = results1[0]
        totalRes2 = results2[0]
        maxSer1 = series1_list[0]
        maxSer2 = series2_list[0]
    else:
        mean1 = sum(results1) / numGames
        mean2 = sum(results2) / numGames
        
        variance1 = sum((x - mean1) ** 2 for x in results1) / (numGames - 1)
        variance2 = sum((x - mean2) ** 2 for x in results2) / (numGames - 1)
        
        sorted1 = sorted(results1)
        sorted2 = sorted(results2)
        median1 = sorted1[numGames // 2] if numGames % 2 == 1 else (sorted1[numGames // 2 - 1] + sorted1[numGames // 2]) / 2
        median2 = sorted2[numGames // 2] if numGames % 2 == 1 else (sorted2[numGames // 2 - 1] + sorted2[numGames // 2]) / 2
        
        from collections import Counter
        counter1 = Counter(results1)
        counter2 = Counter(results2)
        mode1 = counter1.most_common(1)[0][0]
        mode2 = counter2.most_common(1)[0][0]
        
        totalRes1 = mean1
        totalRes2 = mean2
        maxSer1 = max(series1_list)
        maxSer2 = max(series2_list)
    
    points[chel1] += totalRes1
    points[chel2] += totalRes2
    
    if maxSer1 > serie[chel1]:
        serie[chel1] = maxSer1
    if maxSer2 > serie[chel2]:
        serie[chel2] = maxSer2
        
    matrixPoints[chel1][chel2] = totalRes2
    matrixSeries[chel1][chel2] = maxSer2
        
total_pairs = len(strategies) * len(strategies)
current_pair = 0

for i in range(len(strategies)):
    for j in range(len(strategies)):
        current_pair += 1
        chel1 = strategies[i]
        chel2 = strategies[j]
        isStochastic = chel1 in stochasticStrategies or chel2 in stochasticStrategies
        games_count = "10" if isStochastic else "1"
        makeGame(chel1, chel2)

# табліца 1 (ачкі)
results = []
for player, totalPoints in points.items():
    averagePoints = totalPoints / len(strategies)
    results.append([player.__name__, totalPoints, averagePoints])

results.sort(key=lambda x: x[2], reverse=True)

for i, row in enumerate(results, 1):
    row.insert(0, i)

headers = ["Месца", "Стратэгія", "Агульныя ачкі", "Сярэднія ачкі"]
print("Вынікі па найбольшай набранай колькасці ачкоў")
print(tabulate(results, headers=headers, tablefmt="grid"))
print('\n')

# табліца 2 (серыі)
resultsSeries = []
for player, totalSeries in serie.items():
    resultsSeries.append([player.__name__, totalSeries])

resultsSeries.sort(key=lambda x: x[1], reverse=True)

for i, row in enumerate(resultsSeries, 1):
    row.insert(0, i)

headersS = ["Месца", "Стратэгія", "Найбольшая серыя"]
print("Вынікі па найдаўжэйшай серыі выйгрышаў (5 ачкоў)")
print(tabulate(resultsSeries, headers=headersS, tablefmt="grid"))
print('\n')

# табліца 3 (пары)
print("Матрыца ачкоў: колькі выйграў гулец з слупка супраць гульца з радка")
matrixDataPoints = []
for s1 in strategies:
    row = [s1.__name__]
    for s2 in strategies:
        row.append(matrixPoints[s1][s2])
    matrixDataPoints.append(row)

headersMatrix = ["Гулец \\ Супраць"] + strategyNames
print(tabulate(matrixDataPoints, headers=headersMatrix, tablefmt="grid"))
print('\n')

statsResults = []
for player in stochasticStrategies:
    allResults = []
    for opponent in strategies:
        if player != opponent:
            allResults.append(matrixPoints[player][opponent])
    
    if allResults:
        mean_val = sum(allResults) / len(allResults)
        variance_val = sum((x - mean_val) ** 2 for x in allResults) / (len(allResults) - 1)
        sortedResults = sorted(allResults)
        n = len(sortedResults)
        median_val = sortedResults[n // 2] if n % 2 == 1 else (sortedResults[n // 2 - 1] + sortedResults[n // 2]) / 2
        
        from collections import Counter
        counter = Counter(all_results)
        mode_val = counter.most_common(1)[0][0]
        
        statsResults.append([
            player.__name__,
            f"{mean_val:.1f}",
            f"{variance_val:.1f}",
            f"{median_val:.1f}",
            f"{mode_val:.1f}"
        ])

# табліца 4 (статыстыка)
statsResults.sort(key=lambda x: float(x[1]), reverse=True)

for i, row in enumerate(stats_results, 1):
    row.insert(0, i)

headers_stats = ["Месца", "Стратэгія", "Сярэдняе", "Дысперсія", "Медыяна", "Мода"]
print("Аналіз вынікаў стахастычных стратэгій")
print(tabulate(statsResults, headers=headers_stats, tablefmt="grid"))
print('\n')
```
