```python
import math
import matplotlib.pyplot as plt
import numpy as np

def modelExp (x0, r):
    res = []
    x = x0
    for i in range (10):
        x = r*x
        res.append(x)
    return res

def modelLogis (x0, r):
    res = []
    x = x0
    for i in range (50):
        x = r*x*(1-x)
        res.append(x)
    return res

def modelMoran (x0, r):
    res = []
    x = x0
    for i in range (100):
        x = x*math.exp(r * (1 - x))
        res.append(x)
    return res

def modelNicholsonBailey(x0, y0, a, b, c):
    hosts = []    
    parasites = []
    x = x0
    y = y0
    for i in range(50):
        hosts.append(x)
        parasites.append(y)
        x_old = x
        y_old = y
        x = b * x_old * math.exp(-a * y_old)     
        y = c * x_old * (1 - math.exp(-a * y_old)) 
    return hosts, parasites

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

#  
rValuesExp = [0.8, 1.0, 1.2]
for i, r in enumerate(rValuesExp):
    res = modelExp(1, r)
    axes[0, 0].plot(res[:100], label=f'r = {r}')
axes[0, 0].set_title(' ')
axes[0, 0].set_xlabel('')
axes[0, 0].set_ylabel('')
axes[0, 0].legend()
axes[0, 0].grid(True)

#  
rValuesLogis = [ 3.9 ]
for i, r in enumerate(rValuesLogis):
    res = modelLogis(0.5, r)
    axes[0, 1].plot(res[:100], label=f'r = {r}')
axes[0, 1].set_title(' ')
axes[0, 1].set_xlabel('')
axes[0, 1].set_ylabel('')
axes[0, 1].legend()
axes[0, 1].grid(True)

#  
rValuesMoran = [1.5, 2.0, 2.5, 3.0]
for i, r in enumerate(rValuesMoran):
    res = modelMoran(0.7, r)
    axes[1, 0].plot(res, label=f'r = {r}')
axes[1, 0].set_title(' ')
axes[1, 0].set_xlabel('')
axes[1, 0].set_ylabel('')
axes[1, 0].legend()
axes[1, 0].grid(True)

#  -
rValuesNb = [0.05, 0.1, 0.15, 0.2]  #    a
for i, a in enumerate(rValuesNb):
    hosts, parasites = modelNicholsonBailey(50, 10, a, 1.5, 1.0)
    axes[1, 1].plot(hosts[:200], label=f'a = {a}')
axes[1, 1].set_title(' - ()')
axes[1, 1].set_xlabel('')
axes[1, 1].set_ylabel(' ')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

#  
plt.figure(figsize=(12, 8))

for i, a in enumerate(rValuesNb):
    hosts, parasites = modelNicholsonBailey(50, 10, a, 1.5, 1.0)
    plt.plot(parasites[:200], label=f'a = {a}')

plt.title(' - ()')
plt.xlabel('')
plt.ylabel(' ')
plt.legend()
plt.grid(True)
plt.show()


```
