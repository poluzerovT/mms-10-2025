```python
import random
import math
from abc import ABC, abstractmethod
Mat = [
    [3, 0],
    [5, 1]
]
REPEAT = 100000
```

```python
# базовый класс
class p(ABC):
    def __init__(self):
        self._score = 0  #хранит сумму очков
        self._myHistory = []  # хранит историю моих ходов (1/0)
        
    def getScore(self):
        return self._score
    def addScore(self, i):
        self._score += i
    def getHistory(self):
        return self._myHistory
        
    def reload (self): #Сброс настроек
        self._score = 0;
        self._myHistory = []

    @abstractmethod
    def next(self, opHistory):
        pass        
    @abstractmethod
    def name(self):
        pass
```

```python
class pNull(p):
    def next (self, opHist):   #сделать ход
        self._myHistory.append(0)
        return 0
    def name(self):
        return "NULL"
```

```python
class pOne(p):
    def next (self, opHist):   #сделать ход
        self._myHistory.append(1)
        return 1
    def name(self):
        return "ONE"
```

```python
class pRepeat(p):
    def next (self, opHist):   #сделать ход
        if len(self._myHistory) == 0:
            self._myHistory.append(0)
            return 0
        else:
            try:
                a = opHist[len(self._myHistory)-1]
                self._myHistory.append(a)
                return a
            except Exception as e:
                print(len(self._myHistory))
                print(f"{self._myHistory}")
    def name(self):
        return "Repeater"
```

```python
class pOposite(p):
    def next (self, opHist):   #сделать ход
        if len(self._myHistory) == 0:
            self._myHistory.append(0)
            return 0
        else:
            a = 1 if opHist[len(self._myHistory)-1] == 0 else 0
            self._myHistory.append(a)
            return a
    def name(self):
        return "Opositer"
```

```python
class p20_1(p):
    def next (self, opHist):   #сделать ход
        if (len(self._myHistory) + 1) % 20 == 0:
            self._myHistory.append(1)
            return 1
        else:
            self._myHistory.append(0)
            return 0
    def name(self):
        return "20One"
```

```python
class pHurt(p):
    def __init__(self):
        self.__flag = 0
        super().__init__()
    def next (self, opHist):   #сделать ход
        if len(self._myHistory) == 0:
            self._myHistory.append(0)
            return 0
        if opHist[len(self._myHistory)-1] == 1:
            self.__flag = 1
        self._myHistory.append(self.__flag)
        return self.__flag                
    def name(self):
        return "Hurter"
    def reload(self):
        self.__flag = 0
        super().reload()
```

```python
class pMyDet(p):
    def __init__(self):
        self.__coop = 0
        super().__init__()
    def next(self, opHist):
        if len(self._myHistory) == 0:
            self._myHistory.append(0)
            return 0;
        elif opHist[len(self._myHistory)-1] == 0:
            self.__coop += 1
        rate = self.__coop / len(self._myHistory)
        if rate > 0.90:
            self._myHistory.append(0)
            return 0
        elif rate < 0.50:
            self._myHistory.append(1)
            return 1
        else:
            a = opHist[len(self._myHistory)-1]
            self._myHistory.append(a)
            return a
    def name(self):
        return "MyDetermind"
    def reload(self):
        self.__coop = 0
        super().reload()
```

```python
class p05(p):
    def next (self, opHist):
        if random.random() < 0.5:
            self._myHistory.append(0)
            return 0
        else:
            self._myHistory.append(1)
            return 1
    def name(self):
        return "50/50"
    
```

```python
class p09(p):
    def next (self, opHist):
        if random.random() < 0.9:
            self._myHistory.append(0)
            return 0
        else:
            self._myHistory.append(1)
            return 1
    def name(self):
        return "90/10"
```

```python
class pPardon025(p):
    def next (self, opHist):
        if len(self._myHistory) == 0:  #первый ход
            self._myHistory.append(0)
            return 0
        else:  #в зависимости от противника
            if opHist[len(self._myHistory)-1] == 0:  #противник походил 0
                self._myHistory.append(0)
                return 0
            else:#противник походил 1
                if random.random() < 0.25:
                    self._myHistory.append(0)
                    return 0
                else:
                    self._myHistory.append(1)
                    return 1
    def name(self):
        return "Pardon25%"
```

```python
class pNegative025(p):
    def next (self, opHist):
        if len(self._myHistory) == 0:  #первый ход
            self._myHistory.append(0)
            return 0
        else:  #в зависимости от противника
            if random.random() < 0.25:
                a = 1 if opHist[len(self._myHistory)-1] == 0 else 0
                self._myHistory.append(a)
                return a
            else:
                a = opHist[len(self._myHistory)-1]
                self._myHistory.append(a)
                return a
    def name(self):
        return "Negative25%"
```

```python
class pOneIn50(p):
    def __init__(self):
        self.__i = random.randint(1, 50)
        super().__init__()
    def next (self, opHist):
        if (len(self._myHistory)+1) % self.__i == 0:  #i-й ход
            self._myHistory.append(1)
            return 1
        else: 
            self._myHistory.append(0)
            return 0
    def name(self):
        return "OneIn50"
    def reload(self):
        self.__i = random.randint(1, 50)
        super().reload()
```

```python
class pChange20(p):
    def __init__(self):
        self.__flag = 0
        self.__i = random.randint(0, 20)
        super().__init__()
    def next (self, opHist):
        while self.__i <= 0:  #определение флага
            self.__flag = 1 if self.__flag == 0 else 0
            self.__i = random.randint(0, 20)
        self.__i -= 1
        self._myHistory.append(self.__flag)
        return self.__flag
    def name(self):
        return "ChangeIn20"
    def reload(self):
        self.__flag = 0
        self.__i = random.randint(0, 20)
        super().reload()
```

```python
class pEmotional(p):
    def __init__(self):
        self.__mood = 0.5  # 0-1: 0 - злой, 1 - довольный
        self.__forgive = 0.1  # скорость прощения
        self.__sens = 0.3  # чувствительность к предательствам
        super().__init__()
    
    def next(self, opHist):
        if len(self._myHistory) == 0: 
            self._myHistory.append(0)
            return 0
        # Обновляем настроение на основе последнего хода оппонента
        if opHist[len(self._myHistory) -1] == 0:  # сотрудничал
            self.__mood = min(1.0, self.__mood + self.__forgive)
        else:  # предал
            self.__mood = max(0.0, self.__mood - self.__sens)
        
        # Вероятность сотрудничества зависит от настроения
        # + 10% случайности для непредсказуемости и шанса на прощение
        
        if random.random() < self.__mood * 0.9 + random.random() * 0.1:
            move = 0
        else:
            move = 1
        self._myHistory.append(move)
        return move
    
    def name(self):
        return "Emotional"
    
    def reload(self):
        self.__mood = 0.5
        super().reload()
```

```python
def middle(sc):
    sum = 0
    for i in range(len(sc)):
        sum += sc[i]
    return sum / len(sc)
```

```python
def disp (sc):
    c = middle(sc)
    sum = 0
    for i in range (len(sc)):
        sum += (sc[i] - c)**2
    return sum / (len(sc) - 1)
```

```python
def median (sc):
    sc.sort()
    if len(sc) % 2 == 0:
        i = len(sc) // 2
        return (sc[i] + sc[i-1]) / 2
    else:
        i = (len(sc) - 1) // 2
        return sc[i]        
```

```python
def moda (sc):
    sc.sort()
    n = len(sc)               #мощность выборки
    k = int(math.log2(n + 1)) #количество интервалов
    n0 = (sc[-1] - sc[0]) / k     #длина интервала
    max_ind = 0               #индекс максимального интервала
    max_inter = 0             #количество чисел в макс интервалее
    l = sc[0]
    r = l + n0
    j = 0
    inter = 0
    ind = 0
    while j < len(sc):
        if l <= sc[j] <= r:
            inter += 1
            j += 1
        else:
            if inter > max_inter:
                max_inter = inter
                inter = 0
                max_ind = ind
            ind += 1
            l = r
            r = l + n0
    l = sc[0] + max_ind * n0
    r = l + n0
    sum = 0
    a = 0
    for i in range (len(sc)):
        if l <= sc[i] <= r:
            sum += sc[i]
            a += 1
    return sum / a
            
```

```python
PlayersX = [
    pNull(),
    pOne(),
    pRepeat(),
    pOposite(),
    p20_1(),
    pHurt(),
    pMyDet(),
    
    p05(),
    p09(),
    pPardon025(),
    pNegative025(),
    pOneIn50(),
    pChange20(),
    pEmotional()
]
PlayersY = [
    pNull(),
    pOne(),
    pRepeat(),
    pOposite(),
    p20_1(),
    pHurt(),
    pMyDet(),
    
    p05(),
    p09(),
    pPardon025(),
    pNegative025(),
    pOneIn50(),
    pChange20(),
    pEmotional()
]
results_mid = [[_ for _ in range(len(PlayersX))] for _ in range(len(PlayersY))]
results_disp = [[_ for _ in range(len(PlayersX))] for _ in range(len(PlayersY))]
results_median = [[_ for _ in range(len(PlayersX))] for _ in range(len(PlayersY))]
results_moda = [[_ for _ in range(len(PlayersX))] for _ in range(len(PlayersY))]
ScoresX = []
ScoresY = []
```

```python
for i in range (len(PlayersY)):
    for j in range(i, len(PlayersX)):
        for _ in range (REPEAT):
            for _ in range(200):
                a = PlayersY[i].next(PlayersX[j].getHistory())
                b = PlayersX[j].next(PlayersY[i].getHistory())
                PlayersY[i].addScore(Mat[a][b])
                PlayersX[j].addScore(Mat[b][a])
            ScoresX.append(PlayersX[j].getScore())
            ScoresY.append(PlayersY[i].getScore())
            PlayersX[j].reload()
            PlayersY[i].reload()
        results_mid[i][j] = middle(ScoresY)
        results_mid[j][i] = middle(ScoresX)
        results_disp[i][j] = disp(ScoresY)
        results_disp[j][i] = disp(ScoresX)
        results_median[i][j] = median(ScoresY)
        results_median[j][i] = median(ScoresX)
        results_moda[i][j] = moda(ScoresY)
        results_moda[j][i] = moda(ScoresX)
        ScoresX = []
        ScoresY = []
```

```python
print("Средние значения (набор очков игроком слева):")
print(f" "*11, end=" ")
for i in range(len(PlayersX)):
    print(f"{PlayersX[i].name():>11}", end=" ")
print()
for i in range(len(PlayersY)):
    print(f"{PlayersY[i].name():>11}", end="|")
    for j in range(len(PlayersX)):
        print(f"{results_mid[i][j]:>11.0f}", end=" ")
    print()
print("Результаты")
for i in range(len(PlayersY)):
    sum = 0
    for j in range(len(PlayersX)):
        sum += results_mid[i][j]
    print(f"{PlayersY[i].name():>12} have {sum:.0f} points")
```
