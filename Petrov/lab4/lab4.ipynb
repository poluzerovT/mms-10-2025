```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider
```

```python
def exponential (X0_int, t, r):
    arr = [_ for _ in range(t+1)]
    arr[0] = X0_int
    for i in range(t):
        arr[i+1] = r * arr[i]
        
    x = [i for i in range(t+1)]
    plt.figure(figsize=(7, 4))
    plt.plot(x, arr, label="Exponential")
    plt.title("Exponential")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.show()
```

```python
def logical (X0_01, t, r):
    arr = [_ for _ in range(t+1)]
    arr[0] = X0_01
    for i in range(t):
        arr[i+1] = r * arr[i] * (1-arr[i])
        if arr[i+1] < 0 :
            arr[i+1] = 0
        if arr[i+1] >1:
            arr[i] = 1
    x = [i for i in range(t+1)]
    plt.figure(figsize=(7, 4))
    plt.plot(x, arr, label="Logical")
    plt.title("Logical")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.show()
```

```python
def Moran (X0_01, t, r):
    arr = [_ for _ in range(t+1)]
    arr[0] = X0_01
    for i in range(t):
        arr[i+1] = arr[i] * np.exp(r * (1-arr[i]))
    x = [i for i in range(t+1)]
    plt.figure(figsize=(7, 4))
    plt.plot(x, arr, label="Moran")
    plt.title("Moran")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.show()
```

```python
def Parazit (X0_int, Y0_int, t, a, b, c):
    arr = [0] * (t+1)
    y = [0]*(t+1)
    arr[0] = X0_int
    y[0] = Y0_int
    for i in range(t):
        arr[i+1] = round(b * arr[i] * np.exp((-a) * y[i]))
        if arr[i+1] < 0:
            arr[i+1]=0
        # print(arr[i+1], end = " ")
        y[i+1] = round(c * arr[i] * (1 - np.exp(-a * y[i])))
        if y[i+1] < 0:
            y[i+1] = 0
    # x = [i for i in range(t+1)]
    plt.figure(figsize=(7, 4))
    plt.plot(arr, 'b-', label='Хозяева', linewidth=2)
    plt.plot(y, 'r-', label='Паразиты', linewidth=2)
    plt.xlabel('Время')
    plt.ylabel('Популяция')
    plt.title('Устойчивые колебания')
    plt.legend()

    plt.tight_layout()
    
    plt.show()
```

```python
interact(exponential,
         X0_int=IntSlider(min=2, max=10, step=1, value=5, description='Начальная популяция'),
         t=IntSlider(min=10, max=100, step=10, value=50, description='Время жизни'),
         r=FloatSlider(min=0, max=2, step=0.5, value=1, description='Коэффициент роста'))
```

```python
interact(logical,
         X0_01=FloatSlider(min=0.1, max=1, step=0.1, value=0.5, description='Начальная популяция'),
         t=IntSlider(min=10, max=100, step=10, value=50, description='Время жизни'),
         r=FloatSlider(min=0, max=5, step=0.1, value=1, description='Коэффициент роста'))
```

```python
interact(Moran,
         X0_01=FloatSlider(min=0.1, max=1, step=0.1, value=0.1, description='Начальная популяция'),
         t=IntSlider(min=10, max=100, step=10, value=50, description='Время жизни'),
         r=FloatSlider(min=0, max=3, step=0.1, value=1, description='Коэффициент роста'))
```

```python
interact(Parazit,
         X0_int=IntSlider(min=2, max=50, step=5, value=30, description='Хозяинов в начале'),
         Y0_int=IntSlider(min=2, max=50, step=5, value=8, description='Паразитов в начале'),
         t=IntSlider(min=10, max=50, step=5, value=20, description='Время жизни'),
         a=FloatSlider(min=0.001, max=0.1, step=0.01, value=0.05, description='a'),
         b=FloatSlider(min=1.01, max=5, step=0.2, value=1.8, description='b'),
         c=FloatSlider(min=0.1, max=5, step=0.2, value=1.2, description='c'))
```
